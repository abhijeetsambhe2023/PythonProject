```python
df = pd.read_csv('/content/Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df
```

```python
pd.isnull(df).sum()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df=df.dropna()
```

```python
ax=sns.countplot(x='Gender',data=df)
for x in ax.containers:
  ax.bar_label(x)
```

```python
df['Gender'].value_counts()
```

```python
# Plotting barchart for gender vs total amount
sales= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales
```

```python
sns.barplot(x='Gender',y='Amount',data=sales)
```

```python
sales= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='State',y='Amount',data=sales)
```

```python
sales_state= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
sales_state
```

```python
sns.barplot(x='State',y='Amount',data=sales_state)
```
